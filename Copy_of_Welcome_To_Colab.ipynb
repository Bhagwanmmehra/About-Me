```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df=sns.load_dataset("penguins")
df
```

```
df["species"].unique()
```

```
order_a=["Biscoe","Dream","Torgersen"]
sns.barplot(x="island", y="bill_depth_mm", data=df, hue="sex", legend="auto", order=order_a, hue_order=["Female","Male"])
plt.show()
```

```
m={'Female': "o",'Male':">"}
sns.scatterplot(x="bill_depth_mm", y="bill_length_mm",data=df, hue="sex", style="sex", sizes=(30,300), markers=m)

plt.show()
```

```
# c=["r","g"]
sns.countplot( x="species",data=df, hue='sex', palette=("Paired"))
```

```
sns.pairplot(data=df, hue="sex", )
```

```
sns.stripplot(x="species", y="bill_depth_mm", data=df, jitter=True, hue="sex", palette=("Set1"))
```

```
sns.catplot(x="species", y="bill_depth_mm", data=df, kind="swarm", hue="sex", palette=("Set2"))
```

```
sns.boxplot(x="species", y="bill_depth_mm", data=df, hue="sex", palette=("Set3"))
```
